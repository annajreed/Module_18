```python
import pandas as pd
```

```python
#read csv files
df1 = pd.read_csv('202205-citibike-tripdata.csv', low_memory=False)
df2 = pd.read_csv('202305-citibike-tripdata.csv', low_memory=False)
df3 = pd.read_csv('JC-202205-citibike-tripdata.csv', low_memory=False)
df4 = pd.read_csv('JC-202305-citibike-tripdata.csv', low_memory=False)
```

```python
#drop null values
df1.dropna(how='all')
df2.dropna(how='all')
df3.dropna(how='all')
df4.dropna(how='all')
```

```python
#concat nyc data into one dataframe
nyc = [df1, df2]
nycdf = pd.concat(nyc)
```

```python
#concat JC data into one dataframe
jc = [df3, df4]
jcdf = pd.concat(jc)
```

```python
# change time_started/ended columns to datetime
nycdf['time_started'] = pd.to_datetime(nycdf['time_started'])
nycdf['time_ended'] = pd.to_datetime(nycdf['time_ended'])
jcdf['time_started'] = pd.to_datetime(jcdf['time_started'])
jcdf['time_ended'] = pd.to_datetime(jcdf['time_ended'])
```

```python
# concat two files together
dflist = [jcdf, nycdf]
citibikedf = pd.concat(dflist)
```

```python
# add duration column with minutes biked
citibikedf['Duration'] = ((citibikedf.time_ended) - (citibikedf.time_started))/pd.Timedelta(minutes=1)
```
